4. Count the number of duplicates in the Google Play dataset. You must build a function to loop through the Google Play dataset

````python
# Solution
def count_number_duplicates(google_play_data):
    apps_unicas = []
    apps_duplicadas = []
    
    for fila in google_play_data[1:]:
        nombre = fila[0]
        if nombre in apps_unicas:
            apps_duplicadas.append(nombre)
        else:
            apps_unicas.append(nombre)
    num_duplicadas = len (apps_duplicadas)
    return num_duplicadas
numero_duplicadas = count_number_duplicates(google_play_data)
numero_duplicadas
````

````python
# remove duplicates
def quitar_duplicadas(google_play_data):
    revisadas = set()
    apps_unicas = []
    
    apps_unicas.append(google_play_data[0])
    
    for fila in google_play_data[1:]:
        nombres = fila[0]
        if nombres not in revisadas:
            apps_unicas.append(fila)
            revisadas.add(nombres)
    return apps_unicas
google_play_data1 = quitar_duplicadas(google_play_data)
print(len(google_play_data1))
````

5. a. Create a dictionary, where each dictionary key is a unique app name and the corresponding dictionary value is the highest number of reviews of that app


````python
# Solution
def crear_diccionario(google_play_data1):
    diccionario = {}
    
    for fila in google_play_data1[1:]:
        nombres_app = fila[0]
        num_reviews = fila[3] 
        
        if nombres_app in diccionario:
            if num_reviews > diccionario[nombres_app]:
                diccionario[nombres_app] = num_reviews
        else:
            diccionario[nombres_app] = num_reviews
    return diccionario

nuevo_diccionario = crear_diccionario(google_play_data1)

print(nuevo_diccionario)
    
    
````

5. b. Use the information stored in the dictionary and create a new dataset, which will have only one entry per app (and for each app, we'll only select the entry with the highest number of reviews).


````python
def crear_dataset(google_play_data1, nuevo_diccionario):
    nuevo_google_play_data = [google_play_data1[0]]

    for fila in google_play_data1[1:]:
        nombres_app = fila[0]
        num_reviews = fila[3]

        if num_reviews == nuevo_diccionario[nombres_app]:
            nuevo_google_play_data.append(fila)
    return nuevo_google_play_data

nuevo_google_play_data = crear_dataset(google_play_data1, nuevo_diccionario)

nuevo_google_play_data
````

 8. Build a frequency table for the ```prime_genre``` column of the App Store dataset, and the ```Genres``` and ```Category``` columns of the Google Play data set. What are the most common genres for each market? You must develop and deploy your function

````python
# Solution
from collections import defaultdict

def generate_frequency_table(apple_store_data, column_index):
    frequency_table = defaultdict(int)

    for row in apple_store_data[1:]:
        value = row[column_index]
        frequency_table[value] += 1

    return frequency_table

def get_most_common_genre(frequency_table):
    max_frequency = max(frequency_table.values())
    most_common_genres = [key for key, value in frequency_table.items() if value == max_frequency]

    return most_common_genres, max_frequency

apple_store_genre_index = apple_store_data[0].index('prime_genre')
apple_store_frequency_table = generate_frequency_table(apple_store_data, apple_store_genre_index)
apple_store_most_common_genres, apple_store_max_frequency = get_most_common_genre(apple_store_frequency_table)

apple_store_frequency_table
apple_store_most_common_genres, apple_store_max_frequency
````

 9. One way to find out what genres are the most popular (have the most users) is to calculate the average number of installs for each app genre. Based on that exercise, give one app profile recommendation for Google Play.
 
    For the Google Play data set, we can find the required information in the ```Installs``` column. Then, compute the average number of installs by genre. Hint 1: consider only unique values of the ```Category``` column. Hint 2: If an app has, for instance, ```1,000,000+``` assume it as ```1,000,000``` installs. Hint 3: Be careful with a skewed distribution

````python
# Solution
def calculate_average_installs_by_genre(nuevo_google_play_data, category_index, installs_index):
    genre_installs = {}

    for row in nuevo_google_play_data[1:]:
        category = row[category_index]
        installs = row[installs_index].replace('+', '').replace(',', '')  # Quitamos los + y comas de la columna install 
        installs = float(installs)  # Pasamos a formato float

        if category in genre_installs:
            genre_installs[category][0] += installs
            genre_installs[category][1] += 1
        else:
            genre_installs[category] = [installs, 1]

    average_installs_by_genre = {category: genre_installs[category][0] / genre_installs[category][1] for category in genre_installs}
    return average_installs_by_genre

# Calcular la media de instalaciones por genero en Google Play Store dataset
google_play_category_index = nuevo_google_play_data[0].index('Category')
google_play_installs_index = nuevo_google_play_data[0].index('Installs')
average_installs_by_genre = calculate_average_installs_by_genre(nuevo_google_play_data, google_play_category_index, google_play_installs_index)

# Ordenamos de mayor a menos 
sorted_average_installs = sorted(average_installs_by_genre.items(), key=lambda x: x[1], reverse=True)
sorted_average_installs
````

 10. The App Store dataset doesn¬¥t have the ```installs``` information. Then, you will construct the average rating from the ```rating_count_tot``` column per genre. Based on that exercise, give one app profile recommendation for the App Store

````python
# Solution
def calculate_average_rating_by_genre(apple_store_data, genre_index, rating_count_index):
    genre_ratings = {}

    for row in apple_store_data[1:]:
        genre = row[genre_index]
        rating_count = float(row[rating_count_index])  # Convert rating count to float

        if genre in genre_ratings:
            genre_ratings[genre][0] += rating_count
            genre_ratings[genre][1] += 1
        else:
            genre_ratings[genre] = [rating_count, 1]

    average_rating_by_genre = {genre: genre_ratings[genre][0] / genre_ratings[genre][1] for genre in genre_ratings}
    return average_rating_by_genre

# Calculate average rating by genre (prime_genre column) in the App Store dataset
app_store_genre_index = apple_store_data[0].index('prime_genre')
app_store_rating_count_index = apple_store_data[0].index('rating_count_tot')
average_rating_by_genre = calculate_average_rating_by_genre(apple_store_data, app_store_genre_index, app_store_rating_count_index)

# Sort the average rating by genre in descending order
sorted_average_ratings = sorted(average_rating_by_genre.items(), key=lambda x: x[1], reverse=True)
sorted_average_ratings
````
